```python
import numpy as np
import pandas as pd
import time
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import umap
from umap import UMAP
from sklearn import preprocessing
import warnings
warnings.filterwarnings('ignore')

data=pd.read_csv('fashion-mnist_test.csv')
data = data.reset_index()
```

```python
D = data.copy()
scaler = preprocessing.MinMaxScaler()
D = pd.DataFrame(scaler.fit_transform(D), columns=D.columns)
```

```python
T = TSNE(n_components=2, perplexity=5, random_state=123)
TSNE_features = T.fit_transform(D)
```

```python
DF = data.copy()
DF['x'] = TSNE_features[:, 0]
DF['y'] = TSNE_features[:, 1]
```

```python
time_started= time.time()
lsne = TSNE(n_iter=300)
lsne_results = lsne.fit_transform(D)
print('Время работы алгоритма: ', format(round(time.time() - time_started,4)))
fig = plt.figure()
plt.title('perplexity=5')
sns.scatterplot(x='x', y='y', data=DF, hue=data['label'], palette='bright', legend=False)
plt.show()
```

```python
D = data.copy()
scaler = preprocessing.MinMaxScaler()
D = pd.DataFrame(scaler.fit_transform(D), columns=D.columns)
time_started1 = time.time()
umap = UMAP(n_components=2, n_neighbors=5, min_dist=0.1, random_state=123)
umap_feathures = umap.fit_transform(data.iloc[:,1:])
print('время работы алгоритма = ', format(round(time.time() - time_started1,4)))
data['x1'] = umap_feathures[:,0]
data['y1'] = umap_feathures[:,1]
plt.title('n_neighbors=5, min_dist = 0.1')
sns.scatterplot(x = data['x1'], y=data['y1'], hue=data['label'], palette = 'muted',legend=False)
plt.show()
```
