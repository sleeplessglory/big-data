```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as sts
import random
```

```python
data=pd.read_csv('insurance.csv')
```

```python
data.describe()
```

```python
sns.displot(data['age'])
sns.displot(data['bmi'])
sns.displot(data['children'])
sns.displot(data['charges'])
```

```python
mean=np.mean(data['bmi'])
moda=sts.mode(data['bmi'], keepdims=True)
med=np.median(data['bmi'])
std=data['bmi'].std()
raz=data['bmi'].max()-data['bmi'].min()
iqr=sts.iqr(data['bmi'])
```

```python
mean_q=np.mean(data['charges'])
moda_q=sts.mode(data['charges'], keepdims=True)
med_q=np.median(data['charges'])
std_q=data['charges'].std()
raz_q=data['charges'].max()-data['charges'].min()
iqr_q=sts.iqr(data['charges'])
```

```python
print('Среднее =', mean)
print('Мода =', moda)
print('Медиана =', med)
print('Стандартное отклонение =', std)
print('Размах =', raz)
print('Межквартальный размах =', iqr)
print('Среднее =', mean_q)
print('Мода =', moda_q)
print('Медиана =', med_q)
print('Стандартное отклонение =', std_q)
print('Размах =', raz_q)
print('Межквартальный размах =', iqr_q)
```

```python
plt.axvline(mean,color='red')
plt.axvline(moda.mode,color='yellow')
plt.axvline(med,color='black')
plt.legend(labels=['mean', 'moda', 'med'])
plt.hist(data['bmi'])
```

```python
plt.axvline(mean_q,color='red')
plt.axvline(moda_q.mode,color='yellow')
plt.axvline(med_q,color='black')
plt.legend(labels=['mean_q', 'moda_q', 'med_q'])
plt.hist(data['charges'])
```

```python
fig, ax=plt.subplots()
ax.set_title('age')
plt.boxplot(data['age'])
plt.show()

fig, ax=plt.subplots()
ax.set_title('bmi')
plt.boxplot(data['bmi'])
plt.show()

fig, ax=plt.subplots()
ax.set_title('charges')
plt.boxplot(data['charges'])
plt.show()

fig, ax=plt.subplots()
ax.set_title('children')
plt.boxplot(data['children'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import statsmodels.api as sm
```

```python
inshuranceData = data
bmiData = inshuranceData['bmi']
means32 = []
means128 = []
means512 = []
for i in range(300):
    sample = bmiData.sample(32)
    means32.append(sample.mean())
    sample = bmiData.sample(128)
    means128.append(sample.mean())
    sample = bmiData.sample(512)
    means512.append(sample.mean())
fig, ax = plt.subplots(1, 3)
ax[0].hist(means32, color='cyan', label='32')
ax[0].set_title('32')
ax[1].hist(means128, color='magenta', label='128')
ax[1].set_title('128')
ax[2].hist(means512, color='yellow', label='512')
ax[2].set_title('512')
print(f"среднее bmi: {bmiData.mean()}")
print("выборка в 32 элементов:")
print(f"\tстандартное отклонение: {np.std(means32)}")
print(f"\tсреднее: {np.mean(means32)}")
print("выборка в 128 элементов:")
print(f"\tстандартное отклонение: {np.std(means128)}")
print(f"\tсреднее: {np.mean(means128)}")
print("выборка в 512 элементов:")
print(f"\tстандартное отклонение: {np.std(means512)}")
print(f"\tсреднее: {np.mean(means512)}")
plt.show()
```

```python
inshuranceData = data
bmi = inshuranceData['bmi']
charges = inshuranceData['charges']
bmi_ci_95 = st.norm.interval(0.95, loc=np.mean(bmi), scale = st.sem(bmi))
charges_ci_95 = st.norm.interval(0.95, loc=np.mean(charges), scale = st.sem(charges))
bmi_ci_99 = st.norm.interval(0.95, loc=np.mean(bmi), scale = st.sem(bmi))
charges_ci_99 = st.norm.interval(0.99, loc=np.mean(charges), scale = st.sem(charges))
print('bmi ci 95%, 99%:', np.round(bmi_ci_95, 5), np.round(bmi_ci_99, 5))
print('charges ci 95%, 99%: ', np.round(charges_ci_95, 5), np.round(charges_ci_99, 5))
```

```python
inshuranceData = data
bmi = inshuranceData['bmi']
charges = inshuranceData['charges']
sm.qqplot(bmi)
sm.qqplot(charges)
plt.show()
stdBmiData = (bmi - bmi.mean())/bmi.std()
stdChargesData = (charges - charges.mean())/charges.std()
print('BMI: ', st.kstest(stdBmiData, 'norm'))
print('Charges: ', st.kstest(stdChargesData, 'norm'))
```
